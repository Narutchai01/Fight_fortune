```python
#@title P1
import random

def showHp_P1(P1Hp):
    print("\nHP Player 1:|",end ="")
    for i in range(P1Hp):
      print("◙",end ="")
    print("|",P1Hp,"%")

def showMana_P1(P1mana):
    print("MANA Player 1:|",end ="")
    for i in range(P1mana):
      print("•",end ="")
    print("|",P1mana,"%")

def attackP1(action,P2Hp,P1mana):
    if action =='A':
        if P1mana >= 10:
          damage = (random.choice([ 5,15,20 ]))*attP1
          P1mana = P1mana -(10*attP1)
        else:
          damage = 0
          print("มานาไม่พอ")
    elif action =='S':
        if P1mana >= 30:
          damage = (random.choice([ 25,30,35 ]))*attP1
          P1mana = P1mana -(30*attP1)
        else:
          damage = 0
          print("มานาไม่พอ")
    elif action == 'U':
        if P1mana >= 40:
          damage = (random.choice([ 40,50,60 ]))*attP1
          P1mana = P1mana -(40*attP1)
        else:
          damage = 0
          print("มานาไม่พอ")
    else:
        print("สกิลไม่ถูกต้อง")
        damage = 0
        P1mana=P1mana
    P2Hp = P2Hp - damage
    print("Damage =",damage)
    showHp_P1(P1Hp)
    showMana_P1(P1mana)
    print("\t\t\t\t\t\t\t")
    showHp_P2(P2Hp)
    showMana_P2(P2mana)
```
